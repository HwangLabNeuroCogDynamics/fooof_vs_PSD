# Fooof

```python
# build regressors
X = sm.add_constant(fooof_df['age'])
print X
```

```python
# which column should we use
for columns in fooof_df.iloc[:,3:]:
    print columns
    
D_Vars = ['PO_EC_slope', 'PO_EC_yint', 'PO_EO_slope', 'PO_EO_yint', 'MF_EC_slope', 'MF_EC_yint', 'MF_EO_slope', 'MF_EO_yint']
```

```python
for columns in D_Vars:
    
    model = sm.OLS(fooof_df[columns], X).fit()
    print(model.summary())
```

```python
for y in D_Vars: 
    sns.lmplot(x='age', y=y, data=fooof_df,
           palette="muted", height=4)
```
