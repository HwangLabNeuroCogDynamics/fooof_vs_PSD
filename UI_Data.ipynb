```python
variables = df.columns[2:6]
print(variables)
```

```python
for var in variables:
    model = '%s ~ age' %var
    print('TESTING:')
    print(model)
    print(smf.ols(formula=model, data=df).fit().summary())
    sns.lmplot(x='age', y=var, data=df,
           palette="muted", height=4)
    print("-------------------------------------------------------")
    print("-------------------------------------------------------")
    print("-------------------------------------------------------")
    print("-------------------------------------------------------")
    print("-------------------------------------------------------")
```

```python
import numpy as np

ages = np.arange(2,26.5,0.5)
agesxcoef = ages * -0.0722 +2.83
print(agesxcoef)
agesxintercept = ages * -0.0689 -9.242
print(agesxintercept)
df = pd.DataFrame()

for i, age in enumerate(ages):
    tdf = pd.DataFrame()
    
    tdf['power'] = agesxintercept[i]-np.log10(np.power(np.arange(0,60,2), agesxcoef[i]))
    tdf['freq'] = np.arange(0,60,2)
    tdf['age'] = age
    df = df.append(tdf)

a=sns.lineplot(x="freq", y="power", hue="age", data=df)
```
