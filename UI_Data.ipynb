```python
for var in variables:
    model = '%s ~ age' %var
    print('TESTING:')
    print(model)
    print(smf.ols(formula=model, data=df).fit().summary())
    sns.lmplot(x='age', y=var, data=df,
           palette="muted", height=4)
    print("-------------------------------------------------------")
    print("-------------------------------------------------------")
    print("-------------------------------------------------------")
    print("-------------------------------------------------------")
    print("-------------------------------------------------------")
```
